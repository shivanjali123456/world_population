```python
# Now we reorder the columns and putting only the population in new dataframe
wp_= wp.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population', '2000 Population',
                               '2010 Population', '2015 Population', '2020 Population', '2022 Population',]].sum()
wp_
```

```python

# Create a bar plot
plt.figure(figsize=(15, 6))
plt.bar(wp['Country/Territory'], wp['World Population Percentage'], color='skyblue')

# Customize the plot
plt.xlabel('Country')
plt.ylabel('Population Percentage (%)')
plt.title('Country Population Percentage Relative to World Population')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability

# Show the plot
plt.tight_layout()  # Ensures all labels fit within the plot area
plt.show()
```

```python
# Create a bar plot
plt.figure(figsize=(15, 6))
plt.bar(wp['Country/Territory'], wp['Area (km²)'], color='blue')

# Customize the plot
plt.xlabel('Country')
plt.ylabel('Area (km²)')
plt.title('Country Area (km²) vs Country')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability

# Show the plot
plt.tight_layout()  # Ensures all labels fit within the plot area
plt.show()
```

```python


# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(wp_, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=.5)

# Customize the plot
plt.title('Population by Year')
plt.xlabel('Year')
plt.ylabel('Continent')

# Show the plot
plt.tight_layout()
plt.show()
```

```python


# Create a pivot table to structure the data for the stacked area chart
#pivot_table = .pivot_table(index='Year', columns='Country', values='Country_Population', aggfunc='sum')

# Create a stacked area chart
plt.figure(figsize=(15, 6))
sns.set_palette('husl')  # Set color palette
plt.stackplot(wp_.index, wp_.values.T, labels=wp_.columns)

# Customize the plot
plt.title('Stacked Area Chart of World Population by Country Over Time')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), title='Country')

# Show the plot
plt.tight_layout()
plt.show()
```

```python


# Create a box plot to compare population distribution by region
plt.figure(figsize=(15, 6))
sns.boxplot(wp_, palette='Set3')

# Customize the plot
plt.title('Box Plot of Population by Region')
plt.xlabel('Country')
plt.ylabel('Population')

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python


# Create a violin plot to compare population distribution by region
plt.figure(figsize=(10, 6))
sns.violinplot(wp_, palette='Set3')

# Customize the plot
plt.title('Violin Plot of Population by Region')
plt.xlabel('Country')
plt.ylabel('Population')

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python

wp.sort_values(by='Rank', ascending=True, inplace=True)

# Create a ranking plot
plt.figure(figsize=(15, 6))
plt.bar(wp['Country/Territory'], wp['Rank'], color='skyblue')

# Customize the plot
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Ranking of Countries by Population Size')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability

# Show the plot
plt.tight_layout()
plt.show()
```

```python


# Group the data by continent and calculate the total population for each continent
continent_population = wp.groupby('Continent')['World Population Percentage'].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))  # Set the figure size (optional)
plt.pie(continent_population, labels=continent_population.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999', '#66b3ff', '#99ff99', '#c2c2f0', '#ffb3e6', '#c2f0c2'])

# Customize the plot (optional)
plt.title('World Population Distribution by Continent')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd


# Load a built-in world dataset from geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))


# Merge the world dataset with your population data using a common identifier (e.g., 'name' or 'iso_a3')
merged_data = world.merge(wp, left_on='name', right_on='Country/Territory')

# Create a choropleth map
plt.figure(figsize=(15, 10))
merged_data.plot(column='2022 Population', cmap='YlOrRd', legend=True, linewidth=0.8, ax=plt.gca())

# Customize the plot
plt.title('Choropleth Map of World Population')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Show the plot
plt.tight_layout()
plt.show()
```
